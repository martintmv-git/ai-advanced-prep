# Seaborn Workshop
### Tips Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset
tips = sns.load_dataset("tips")

# bar plot to show the average total bill for each day, grouped by time
sns.barplot(x="day", y="total_bill", hue="time", data=tips, palette="Set2")

plt.title("Average Total Bill by Day and Time")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill")

plt.show()
```

```python
import seaborn as sns

data = sns.load_dataset("tips")
numeric_data = data.select_dtypes(include=['float64', 'int64'])
corr = numeric_data.corr()
sns.clustermap(corr, cmap="coolwarm")
```

```python
import seaborn as sns

data = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=data)
```

```python
import seaborn as sns

data = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", hue="day", data=data, kind="scatter")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
g = sns.PairGrid(data, hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.show()
```
